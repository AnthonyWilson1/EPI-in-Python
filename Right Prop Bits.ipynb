```python
# Online Python - IDE, Editor, Compiler, Interpreter
# Right propagate the rightmost set bit in x, e.g. turns (01010000) to (01011111).
# 80


def prop_right(x):
    n = (x &~ (x - 1))
    return n

prop_right(80)
```

```python
bit_s = '10000'
  
# convert binary string 
# into integer
temp = int(bit_s, 2)
  
# applying Ex-or operator
# b/w 10 and 31
inverse_s = temp ^ (2 ** (len(bit_s) + 1) - 1) 
# convert the integer result 
# into binary result and then 
# slicing of the '0b1' 
# binary indicator
rslt = bin(inverse_s)
  
# print the result
print("Inversed string is", 
      rslt )
```

```python
#final algo

def prop_right_final(x):
    bitmask = 0xFF
    n = (x &~ (x - 1))
    n_len = len(bin(n))
    index = n_len - 3
    front_mask = (2 ** (index + 1) -1)
    
    return bin(x | front_mask)

prop_right_final(80)

```
