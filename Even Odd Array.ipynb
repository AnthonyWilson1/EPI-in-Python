```python
def even_odd(a):
    even_count = 0
    odd_count = len(a) - 1
    
    while even_count < odd_count:
        if a[even_count] % 2 == 0:
            even_count += 1
            print(even_count)
        else:
            a[odd_count], a[even_count] = a[even_count], a[odd_count]
            odd_count -= 1
            print(odd_count)
        print(a)

arr = [2,2,1,1]   

even_odd(arr)
```

```python
def odd_even(a):
    even_count = 0
    odd_count = len(a) - 1
    
    while even_count < odd_count:
        if a[even_count] % 2 != 0:
            even_count += 1
        else:
            a[odd_count], a[even_count] = a[even_count], a[odd_count]
            odd_count -= 1
    
    return(a)

arr = [2,2,1,1]   

odd_even(arr)
```
