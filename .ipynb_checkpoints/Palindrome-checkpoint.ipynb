```python
import math
```

```python
def palindrome(a):
    #if  number is < 0 and or 0 return
    if a <= 0:
        return a == 0
    
    #length of a
    len_a = math.floor(math.log10(a)) +1
    
    #most sig digit
    sig_dig = 10**(len_a-1)
    
    #loop through and compare the edges
    for i in range(0,len_a//2):
        if a//sig_dig == a % 10:
            
            #slice of left side
            a = a//10
            
            #slice of right side
            a = a % len_a
        
            #reset length
            len_a = len_a//100
    
        else: 
            return False
        
    return True

            
            
```

```python
palindrome(110)
```
