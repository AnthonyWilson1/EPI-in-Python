```python
# Online Python - IDE, Editor, Compiler, Interpreter
# Right propagate the rightmost set bit in x, e.g. turns (01010000) to (01011111).
# 80


def prop_right(x):
    #find the rightmost set bit and find it's index
    #print((bin(x)))
    #print(bin(x &~ (x - 1)))
    #print(len(bin(x))-len(bin(x &~ (x - 1))))
    n = (x &~ (x - 1))
    #everything after has to be 1s & with itself and ~ after the index
    return n

prop_right(80)
```

```python
bit_s = '10000'
  
# convert binary string 
# into integer
temp = int(bit_s, 2)
  
# applying Ex-or operator
# b/w 10 and 31
print(bin((2 ** (len(bit_s) + 1) -1)))
inverse_s = temp ^ (2 ** (len(bit_s) + 1) - 1) 
# convert the integer result 
# into binary result and then 
# slicing of the '0b1' 
# binary indicator
rslt = bin(inverse_s)
  
# print the result
print("Inversed string is", 
      rslt )
```

```python
#bitmask

bitmask = 0xFFFF
x = prop_right(80)
print(bin(x))
print(bin(bitmask))

y = (x ^ (2 ** (len(bin(x))-3 + 1) -1))



print(bin(y | 80))
```
